```python
# Importing all necessary libraries

import pandas as pd
pd.set_option("display.max_columns", None)
from sqlalchemy import create_engine
```

```python
# Importing the data from a csv into a Dataframe 

df = pd.read_csv("C:\\Data Worker Reskilling\\Projects\\TMDB + IMDB Merged Movies Dataset\\data\\raw\\TMDB_IMDB_Movies_Dataset.csv")
```

```python
# Checking if the data was successfully imported

df.head()
```

```python
# Checking if all datatypes were infered correctly by pandas

df.info()
```

```python
# Changing the Datatype of column "release_date" to its correct datatype

df["release_date"] = pd.to_datetime(df["release_date"])
```

```python
# Dropping unnecessary columns

df.drop("overview", axis = 1, inplace = True)
```

```python
# Checking the results of previous operations

df.info()
```

```python
# Getting further info about the values 

df.describe()
```

```python
# Getting info about the NA-values in the dataset

df.isna().sum()
```

```python
# Calculating the percentage of NA-values
# We will continue with the NA-values in this case and decide in the PostgreSQL database if and how they would impact our analysis

df.isna().mean() * 100
```

```python
# Extracting the main genre, main production company, main production country, main spoken language, main director and main writer out of the dataset and checking if operation was successfull

df["main_genre"] = df["genres"].str.split(",").str[0].str.strip()
df["main_production_company"] = df["production_companies"].str.split(",").str[0].str.strip()
df["main_production_country"] = df["production_countries"].str.split(",").str[0].str.strip()
df["main_spoken_language"] = df["spoken_languages"].str.split(",").str[0].str.strip()
df["main_director"] = df["directors"].str.split(",").str[0].str.strip()
df["main_writer"] = df["writers"].str.split(",").str[0].str.strip()
df.head()
```

```python
# Renaming two columns to fit the naming convention

df.rename(columns={"averageRating": "average_rating", "numVotes": "num_votes"}, inplace = True)
```

```python
# Checking if operation was successfull

df.head()
```

```python
# Cheking for a last time, if all datatypes are correct

df.info()
```

```python
# Creating another version of the Dataframe for exporting

movies_complete = df
```

```python
# Exporting the Dataframe with all data in a csv-file and excel-file mainly as a backup

movies_complete.to_csv("C:\\Data Worker Reskilling\\Projects\\TMDB + IMDB Merged Movies Dataset\\data\\processed\\movies_complete.csv", index=False)
movies_complete.to_excel("C:\\Data Worker Reskilling\\Projects\\TMDB + IMDB Merged Movies Dataset\\data\\processed\\movies_complete.xlsx", index=False, engine="openpyxl")
```

```python
# Connecting to the PostgreSQL database and importing the Dataframe into it also mainly as a backup and for ad-hoc analysis

engine = create_engine("postgresql://postgres:12345678@localhost:5432/movies")
movies_complete.to_sql("movies_complete_final", engine, if_exists="replace", index=False)
```

```python
# Continuing to work on the Dataframe by dropping unneccessary columns for analytics

df.drop("vote_average", axis = 1, inplace = True)
df.drop("vote_count", axis = 1, inplace = True)
df.drop("popularity", axis = 1, inplace = True)
df.drop("backdrop_path", axis = 1, inplace = True)
df.drop("homepage", axis = 1, inplace = True)
```

```python
# Checking the success 

df.info()
```
